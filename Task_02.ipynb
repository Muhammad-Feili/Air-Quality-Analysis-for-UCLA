# loading the dataset

```python
import pandas as pd

data=pd.read_csv('NewAirQualityUCI.csv', index_col=0, parse_dates=True)
```

# concatenating the Date and Time columns to discretizing

```python
data['Date']=data['Date']+' '+data['Time']
data.reindex()
data.reset_index(drop=True,inplace=True)
```

# converting the DateTime column to datetime object

```python
import datetime as dt

data['Date']=pd.to_datetime(data['Date'])
data=data.drop('Time',axis=1)
data.rename(columns={'Date':'DateTime'},inplace=True)
```

# dicretizing in 6 hours 
in this operation i use the period index method to produce the alternatice time series in 6 hours

```python
data.DateTime.sort_values()
discretize=pd.PeriodIndex(start='2004-03-11 06:00:00', end='2005-04-04 00:00:00', freq='6H')
discretize[:]
```

# statements in order to discretizing operation

```python
discretize=discretize.to_timestamp()
for i in range(len(data.DateTime)-1):
    j=0
    while data.DateTime[i] > discretize[j]:
        j+=1
    data.DateTime[i]=discretize[j]
```

# sequential pattern mining by pymining library and minimum length of 4

```python
from pymining import seqmining
freq_seqs = seqmining.freq_seq_enum(data, 4)
sorted(freq_seqs)
```


# visualizing the results in some way

```python
from pymining import itemmining, assocrules, perftesting

relim_input = itemmining.get_relim_input(data)
report = itemmining.relim(relim_input, min_support=2)
report
```
