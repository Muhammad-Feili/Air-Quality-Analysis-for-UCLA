# finding the anomalies that is away from mean of which column by distance 4 times of standard deviation that is approximately 1 percent

```python
import numpy as np
import math
import matplotlib.pyplot as plt

anomalies = []

# multiply and add by random numbers to get some real values


# Function to Detection Outlier on one-dimentional datasets.
def find_anomalies(data):
    # Set upper and lower limit to 3 standard deviation
    data_std = np.std(data)
    data_mean = np.mean(data)
    anomaly_cut_off = data_std * 4
    
    lower_limit  = data_mean - anomaly_cut_off 
    upper_limit = data_mean + anomaly_cut_off
    print(lower_limit)
    # Generate outliers
    anomalies={}
    for columns in data:
        anomalies[columns]=[]
        for outlier in data[columns]:
            if outlier > upper_limit[columns] or outlier < lower_limit[columns]:
                anomalies[columns].append(outlier)
                
    return anomalies

anomalies=find_anomalies(data)
```
